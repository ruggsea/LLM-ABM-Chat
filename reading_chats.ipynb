```python
## pretty print the first chat. color the speaker names
import re
def color_names(chat):
    # remove double speaker names, i.e. jasper: jasper: or jasper: vivian: -> jasper: or jasper: vivian: -> jasper:
    chat = re.sub(r"(\w+): \1:", r"\1:", chat)
    chat = re.sub(r"(\w+): (\w+):", r"\1: \2:", chat)
    # alternating two colors
    colors = ["\033[91m", "\033[92m"]
    color_index = 0
    # find all speaker names
    speakers = re.findall(r"(\w+):", chat)
    # remove duplicates
    speakers = list(set(speakers))
    # replace all speaker names with colored versions
    for speaker in speakers:
        chat = chat.replace(speaker + ":", colors[color_index % 2] + speaker + ":\033[0m")
        color_index += 1
    # remove double newlines and double spaces
    chat = re.sub(r"\n\n", "\n", chat)
    chat = re.sub(r"  ", " ", chat)

    return chat

print(color_names(baseline_chats[0]))
```
